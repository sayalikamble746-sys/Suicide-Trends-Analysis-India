```python
import pandas as pd
df =pd.read_csv("Suicides by Educational Status_Sample_Data.csv")
print(df.head());
```

```python
df.columns = [
    "Year",
    "State",
    "State code",
    "Education",
    "Gender",
    "Deaths"
]
print(df);
```

```python
import matplotlib.pyplot as plt
# 1. Education-wise Analysis
import pandas as pd
import matplotlib.pyplot as plt

# Shorten labels (VERY IMPORTANT)
df["Education"] = df["Education"].replace({
    "Middle (Up To Class-8th)": "Middle",
    "Matriculate/ Secondary (Up To Class-10th)": "Secondary",
    "Primary (Up To Class-5th)": "Primary",
    "Hr. Secondary/ Intermediate/ Pre-University (Up To Class-12th)": "Higher Sec",
    "No Education": "No Edu",
    "Status Not Known": "Unknown",
    "Diploma/ Certificate/ Iti": "Diploma",
    "Graduate And Above": "Graduate+",
    "Professionals (Mba Etc.)": "Professional"
})

# Group data
edu = df.groupby("Education")["Deaths"].sum().sort_values(ascending=True)

# Plot (HORIZONTAL BAR - BEST 🔥)
plt.figure(figsize=(10,6))

plt.barh(edu.index, edu.values)

plt.title("Suicides by Education Level", fontsize=14, weight='bold')
plt.xlabel("Total Deaths")
plt.ylabel("Education Level")

plt.tight_layout()
plt.savefig("Suicides by Education Level")
plt.show()
```

```python
# 2. Year-wise Trend
yearly = df.groupby("Year")["Deaths"].sum()

print("\nYear-wise Deaths:\n", yearly)

# Plot
yearly.plot(kind='line', marker='o')
plt.title("Year-wise Suicide Trend")
plt.xlabel("Year")
plt.ylabel("Deaths")
plt.savefig("Year-wise Suicide Trend")
plt.show()
```

```python
# 3. Gender-wise Analysis

gender = df.groupby("Gender")["Deaths"].sum()

print("\nGender-wise Deaths:\n", gender)

# Plot
gender.plot(kind='pie', autopct='%1.1f%%')
plt.title("Gender Distribution")
plt.ylabel("")
plt.savefig("Gender Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))

gender.plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90
)

plt.title("Gender Distribution of Suicide Cases")
plt.ylabel("")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Suicides by Educational Status_Sample_Data.csv")

# Rename columns (adjust if needed)
df.columns = ["Year", "State", "Gender", "Education", "Category", "Deaths"]

# -------------------------------
# 1. Education-wise Analysis
# -------------------------------
edu = df.groupby("Education")["Deaths"].sum().sort_values(ascending=False)

print("\nEducation-wise Deaths:\n", edu)

# Plot
edu.plot(kind='bar')
plt.title("Suicides by Education Level")
plt.xlabel("Education")
plt.ylabel("Deaths")
plt.xticks(rotation=45)
plt.show()

# -------------------------------
# 2. Year-wise Trend
# -------------------------------
yearly = df.groupby("Year")["Deaths"].sum()

print("\nYear-wise Deaths:\n", yearly)

# Plot
yearly.plot(kind='line', marker='o')
plt.title("Year-wise Suicide Trend")
plt.xlabel("Year")
plt.ylabel("Deaths")
plt.show()

# -------------------------------
# 3. Gender-wise Analysis
# -------------------------------
gender = df.groupby("Gender")["Deaths"].sum()

print("\nGender-wise Deaths:\n", gender)

# Plot
gender.plot(kind='pie', autopct='%1.1f%%')
plt.title("Gender Distribution")
plt.ylabel("")
plt.show()

# -------------------------------
# 4. State-wise Analysis (Top 5)
# -------------------------------
state = df.groupby("State")["Deaths"].sum().sort_values(ascending=False).head(5)

print("\nTop 5 States:\n", state)

# Plot
state.plot(kind='bar')
plt.title("Top 5 States by Suicide Cases")
plt.xlabel("State")
plt.ylabel("Deaths")
plt.xticks(rotation=45)
plt.show()
```

```python
 # 4. State-wise Analysis (Top 5)
# -------------------------------
state = df.groupby("State")["Deaths"].sum().sort_values(ascending=False).head(5)

print("\nTop 5 States:\n", state)

# Plot
state.plot(kind='bar')
plt.title("Top 5 States by Suicide Cases")
plt.xlabel("State")
plt.ylabel("Deaths")
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

gender = df.groupby("Gender")["Deaths"].sum()

plt.figure(figsize=(6,6))

gender.plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90
)

plt.title("Gender Distribution")
plt.ylabel("")

plt.show()
```

```python
import matplotlib.pyplot as plt

gender = df.groupby("Gender")["Deaths"].sum()

plt.figure(figsize=(6,4))
gender.plot(kind='bar')

plt.title("Gender-wise Suicide Cases", fontsize=14)
plt.xlabel("Gender")
plt.ylabel("Total Deaths")

plt.xticks(rotation=0)

plt.tight_layout()
plt.savefig("gender_analysis.png")  # fixed file name
plt.show()
```

```python
pivot = df.pivot_table(values="Deaths",
                       index="Education",
                       columns="Gender",
                       aggfunc="sum")

pivot.plot(kind='bar', figsize=(8,5))

plt.title("Education vs Gender Suicide Analysis")
plt.xlabel("Education Level")
plt.ylabel("Total Deaths")

plt.xticks(rotation=30)
plt.tight_layout()
plt.savefig("Education vs Gender Suicide Analysis")
plt.show()
```

```python
trend = df.pivot_table(values="Deaths",
                       index="Year",
                       columns="Gender",
                       aggfunc="sum")

trend.plot(figsize=(7,4), marker='o')

plt.title("Year-wise Trend by Gender")
plt.xlabel("Year")
plt.ylabel("Total Deaths")

plt.grid(True)
plt.tight_layout()
plt.savefig("Year-wise Trend by Gender")
plt.show()
```

```python
edu = df.groupby("Education")["Deaths"].sum()

edu_percent = (edu / edu.sum()) * 100

edu_percent.plot(kind='pie', autopct='%1.1f%%', figsize=(9,6))

plt.title("Percentage Distribution by Education Level")
plt.ylabel("")
plt.tight_layout()
plt.savefig("Percentage Distribution by Education Level")
plt.show()
```

```python
edu_percent.plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(5,4),
    labels=None   # remove long labels
)
plt.legend(edu_percent.index, bbox_to_anchor=(1,1))
plt.savefig("piechart");
```
